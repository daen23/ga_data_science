### <b>Create a table of descriptive statistics for each of the variables (n, mean, median, standard deviation)</b>

```python
#Drop missing datapoints. Apply to dataframe permanently.
df.dropna(inplace=True)

#Create table of basic summary statistics. 
#We could use df.describe(), but this does does not have median, mode, or skew

sumstat = []
for column in df:
    sumstat.append([df[column].count(),
                    df[column].mean(),
                    df[column].median(),
                    df[column].mode()[0],
                    df[column].std(),
                    df[column].skew(),
                    df[column].kurtosis(),
                    df[column].quantile(0.25),
                    df[column].quantile(0.5),
                    df[column].quantile(0.75),
                    df[column].max(),
                    df[column].min()
    ])   

ss = pd.DataFrame(sumstat, index=['admit', 'GRE', 'GPA', 'Prestige'], columns=['Records (exc. nulls)', 'Mean', 'Median', 'Mode', 'Standard Deviation', 'Skew', 'Kurtosis', '25th Percentile', '50th Percentile', '75th Percentile', 'Max', 'Min'])

#transpose so it's easier to read
summarystats = ss.T
summarystats
```

<b><i>Confirm that you dropped the correct data. How can you tell?</i></b><br/>
We can confirm the records with null entries have been dropped. We identified 3 records in previous steps that contained at least one null value in any of the GRE, GPA or Prestige variables. The number of observations is now 397 across all variables, which is consistent.

<b><i>Why would GRE have a larger standard deviation than GPA?</i></b><br/>
It looks like GRE has a larger standard deviation than GPA due to the different units of measurement - it looks like GRE has observations with values between 220 and 800, while GPA has observations with values between 2.26 and 4.0.

Due to GRE's unit of measurement being more granular, students can have a potentially wider range of GRE scores than GPA scores (given that GPA is only measured to 2 decimal points here). In fact we do observe a wider spread of scores in the sample, leading to a higher standard deviation.

### <b>Describe the distributions of your data.</b>

```python
#Set up subplots for Histograms
fig, axes = plt.subplots(1, 3, figsize=(15,5), dpi = 150)
df['gre'].hist(ax= axes[0], grid=False, bins = 15, color='powderblue')
df['gpa'].hist(ax= axes[1], grid=False, bins = 15, color='plum')
df['prestige'].hist(ax= axes[2], grid=False, color='peachpuff')

#Label charts
labels = ['GRE', 'GPA', 'Prestige']
axes[0].set_title('GRE Histogram')
axes[0].set_xlabel('GRE Scores')
axes[1].set_title('GPA Histogram')
axes[1].set_xlabel('GPA Scores')
axes[2].set_title('Prestige Histogram')
axes[2].set_xlabel('Prestige Scores')
for ax in axes:
    ax.set_ylabel('Counts')

plt.tight_layout()
```

<b><i>GRE Observations</i></b><br/>
The GRE data looks normally distributed, with a peak in scores around 550 to 650 (i.e. the mode). This variable probably would fit under an assumption of normal distribution and looks an immediate good fit for a linear regression model without further transformation.

<b><i>GPA Observations</i></b><br/>
The GPA data looks "less normal" than the GRE observations. While the general shape is normal, the mode is right at the top end of the scores (around 4.0).

<b><i>Prestige Observations</i></b><br/>
The Prestige data looks to be normally distributed, with a slight left skew.

### <b>Plot box plots for each variable</b>

```python
#Set up subplots for Histograms, use Seaborn
fig, axes = plt.subplots(1, 3, figsize=(15,5), dpi = 150)
sns.boxplot(df['gre'], ax= axes[0], orient='v', width=0.2, linewidth=2, color = 'powderblue')
sns.boxplot(df['gpa'], ax= axes[1], orient='v', width=0.2, linewidth=2, color = 'plum')
sns.boxplot(df['prestige'], ax= axes[2], orient='v', width=0.2, linewidth=2, color = 'peachpuff')

#Label charts
labels = ['GRE', 'GPA', 'Prestige']
axes[0].set_title('GRE Boxplot')
axes[0].set_ylabel('GRE Scores')
axes[1].set_title('GPA Boxplot')
axes[1].set_ylabel('GPA Scores')
axes[2].set_title('Prestige Boxplot')
axes[2].set_ylabel('Prestige Scores')

plt.tight_layout()
```

<b><i>GRE Boxplot Observations</i></b><br/>
Based on the boxplot, the GRE data only a few outliers, while the rest of the data is concentrated such that we might expect a normal distribution. This, combined with the histogram, should give us confidence that this data should be ready for linear regression without further transformation. 

<b><i>GPA Boxplot Observations</i></b><br/>
The GPA boxplot looks slightly left skewed, given that the 25th and 75th percentile boundaries appear to be towards the higher end of the GPA scores. This boxplot, by itself, is probably not enough to conclude whether the GPA data is normally distributed. 

<b><i>Prestige Observations</i></b><br/>
It's hard to see, but Median for prestige is 2.0. Note that the median coincides with the 25th percentile, suggesting a right skew in the data.

<b>What did you find? Which of our variables are potentially colinear?</b>

Using the correlation matrix, we can see that none of the variables by themselves have a strong correlation with admittance (e.g. none of the variables are a particularly good predictor of admittance by themselves).

It's also interesting that by itself, prestige seems to have a prima facie negative relationship with all other variables, particularly on likelihood of admittance - e.g. all other factors equal, as prestige goes up, probability of admittance goes down. Based on these correlation values, we'd also expect a student's GPA and GRE score to go down as prestige increases. 

In respect of colinearity, it does appear that there is a relationship between GRE and GPA - as GPA increases, so does GRE. It could be a problem as in isolation, GPA and GRE are stronger at explaining each other than admittance, so we'd have find some way to either remove the effect of multicollinearity or find some other variables to replace GRE or GPA.

### <b>Replace missing values using the median replacement method. </b>

```python
#Apply to dataframe permanently, repeat above steps and analysis
df.fillna({'gre': df['gre'].median(), 'gpa': df['gpa'].median(), 'prestige': df['prestige'].median()}, inplace=True)
```

### Replace missing values using multiple imputation methods.
We could replace nulls with other static values - e.g. mean, mode:<br/>

```python
#Mean Replacement Method
df.fillna({'gre': df['gre'].mean(), 'gpa': df['gpa'].mean(), 'prestige': df['prestige'].mean()}, inplace=True)
```

```python
#Mode Replacement Method. 
#Note how we have to index [0] (true mode), as mode of a pandas series can give multiple 'modes'
df.fillna({'gre': df['gre'].mode()[0], 'gpa': df['gpa'].mode()[0], 'prestige': df['prestige'].mode()[0]}, inplace=True)
```

### <b>Log transform data to meet normality requirements</b>

```python
#read in the dataset as a dataframe
df = pd.read_csv("~/Dropbox/GA Data Science Course/Week 1 Materials/admissions.csv")

#Replace missing values using the median replacement method to Tidy things up. Apply to dataframe permanently
df.fillna({'gre': df['gre'].median(), 'gpa': df['gpa'].median(), 'prestige': df['prestige'].median()}, inplace=True)

df1 = pd.DataFrame(data = [df['admit'], np.log(df['gre']), np.log(df['gpa']), np.log(df['prestige'])])
df_norm = df1.T

#Set up subplots for Histograms to check
fig, axes = plt.subplots(1, 3, figsize=(15,5), dpi = 150)
df_norm['gre'].hist(ax= axes[0], grid=False, bins = 15, color='powderblue')
df_norm['gpa'].hist(ax= axes[1], grid=False, bins = 15, color='plum')
df_norm['prestige'].hist(ax= axes[2], grid=False, color='peachpuff')

#Label charts
labels = ['GRE', 'GPA', 'Prestige']
axes[0].set_title('GRE Histogram')
axes[0].set_xlabel('Natural Log of GRE Scores')
axes[1].set_title('GPA Histogram')
axes[1].set_xlabel('Natural Log of GPA Scores')
axes[2].set_title('Prestige Histogram')
axes[2].set_xlabel('Natural Log of Prestige Scores')
for ax in axes:
    ax.set_ylabel('Counts')

plt.tight_layout()
```

Taking the natural log of the data did not make a meaningful change in the shape of the distribution of the data. Also note that since our dependent variable is binary (i.e. 0 or 1), we will not be able to take the natural log without additional transformation of the data (since Ln(0) is undefined).
